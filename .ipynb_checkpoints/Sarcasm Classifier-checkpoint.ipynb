```python
import json

def parse_data(file):
    for n in open(file, 'r'):
        yield json.loads(n)
        
datastore = list(parse_data("dataset/Sarcasm dataset/Sarcasm_Headlines_Dataset_v2.json"))

sentences = []
labels = []
urls = []

for item in datastore:
    sentences.append(item['headline'])
    labels.append(item['is_sarcastic'])
    urls.append(item['article_link'])
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# #tokenize by replacing missing words with <00V>
# tokenizer = Tokenizer(oov_token="<00V>", filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
# tokenizer.fit_on_texts(sentences)
# word_index = tokenizer.word_index

# sequence = tokenizer.texts_to_sequences(sentences)
# padding = pad_sequences(sequence, padding='post')

# print(word_index)
# print(sequence)
# print(padding)
```

# Split the data

```python
print('[INFO] Splitting the data'+'...'*5)

training_size = 28619

train_sentences = sentences[0:training_size]
test_sentences = sentences[training_size:]
train_labels = labels[0:training_size]
test_labels = labels[training_size:]
```

```python
print('[INFO] Tokenizing words'+'...'*5)

vocab_size = 10000
oov_tok='<00V>'
max_length = 100
padding_type = 'post'
trunc_type = 'post'
embedding_dim = 64


tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index

train_sequences = tokenizer.texts_to_sequences(train_sentences)
train_padded = pad_sequences(train_sequences, maxlen=max_length,
                             padding=padding_type, truncating=trunc_type)

test_sequences = tokenizer.texts_to_sequences(test_sentences)
test_padded = pad_sequences(test_sequences, maxlen=max_length,
                            padding=padding_type, truncating=trunc_type)

print(train_sequences)
print(train_padded)
```

```python
#convert this data into numpy array inorder to feed it into the model
import numpy as np
train_padded = np.array(train_padded)
train_labels = np.array(train_labels)

test_padded = np.array(test_padded)
test_labels = np.array(test_labels)
```

# Outline the Model

```python
hidden_layer_size = 350

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length = max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
print('Compiling the Model'+'...'*5)
INIT_lr = 0.005 #initialize the learning rate
opt = tf.keras.optimizers.Adam(INIT_lr)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print('Training the Model'+'...'*5)
num_epoch=10 #initialize number of epochs to train for
History = model.fit(train_padded, train_labels, epochs=num_epoch, 
                   validation_data=(test_padded, test_labels), verbose=2, batch_size=25)
```

```python
print('[INFO] Summary of the Model'+'...'*5)
model.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
N=num_epoch

plt.style.use('ggplot')
plt.figure(figsize=(12,7))
plt.plot(np.arange(0,N), History.history['loss'],'r', lw=2, label='training loss')
plt.plot(np.arange(0,N), History.history['accuracy'],'g', lw=2, label='training accuracy')
plt.title('Training Loss and Accuracy')
plt.xlabel('Number of Epochs')
plt.ylabel('Loss/Accuracy')
plt.legend()
plt.show()
#print('Saving plot'+'...'*5)
# plt.savefig('plot')
```

# Make prediction

```python
print('[INFO] Making predictions'+'...'*5)
sentence = str(input('Enter a sentence: '))
sequence = tokenizer.texts_to_sequences(sentence)
pred_padded = pad_sequences(sequence, maxlen=max_length,
                           padding=padding_type, truncating=trunc_type)
predict = model.predict(pred_padded)
i = predict.shape[0]
#i=89
def answer(s):
    if s >= 50:
        print("This sentence is Sarcastic with: {}".format(s)+"%")
    else:
        print("This sentence is Not Sarcastic with: {}".format(s)+"%")
```
